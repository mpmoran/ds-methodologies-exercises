1. Write a function named is_vowel. It should accept a string as input and use a regular expression to determine if the passed string is a vowel. While not explicity mentioned in the lesson, you can treat the result of re.search as a boolean value that indicates whether or not the regular expression matches the given string.

```python
def is_vowel(s: str) -> bool:
    """
    returns True if s is a vowel; otherwise, False
    """
    return True if re.search(r"^[aeiou]$", s) else False

assert is_vowel("a")
assert not is_vowel("z")
assert is_vowel("e")
assert is_vowel("i")
assert is_vowel("o")
assert is_vowel("u")
assert not is_vowel("aeu")
assert not is_vowel("4")
```

2. Write a function named is_valid_username that accepts a string as input. A valid username starts with a lowercase letter, and only consists of lowercase letters, numbers, or the _ character. It should also be no longer than 32 characters. The function should return either True or False depending on whether the passed string is a valid username.

```python
def is_valid_username(username: str) -> bool:
    """
    
    """
    re_uname = r"^[a-z][a-z0-9_]{0,31}$"
    return True if re.search(re_uname, username) else False

assert not is_valid_username('aaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaa')
assert is_valid_username('codeup')
assert not is_valid_username('Codeup')
assert is_valid_username('codeup123')
assert not is_valid_username('1codeup')
```

**6. You can find a list of words on your mac at /usr/share/dict/words. Use this file to answer the following questions:**

```python
f = open("/usr/share/dict/words", "r")
words = f.read()
f.close()
```

- How many words have at least 3 vowels?

```python
matches = re.findall(r"^[AEIOUaeiou][a-z]*?[aeiou][a-z]*?[aeiou]", words, re.MULTILINE)
print(len(matches))
# print(matches[:10])
for word in matches:
    nvowels = 0
    for ltr in word:
        if ltr in "AEIOUaeiou":
            nvowels += 1
    if nvowels != 3:
        print(f"{repr(word)} has {nvowels} vowels")
```

- How many words have at least 3 vowels in a row?

```python
matches = re.findall(r"[AEIOUaeiou]{3}", words, re.MULTILINE)
print(len(matches))
```

- How many words have at least 4 consonants in a row?

```python
matches = re.findall(r"[^AEOIUaeiou\s\d]{4}", words, re.MULTILINE)
print(len(matches))
print(matches[:10])
```

- How many words start and end with the same letter?

```python
matches = re.findall(r"^([A-Za-z])[A-Za-z]*?\1$", words, re.MULTILINE)
print(len(matches))
print(matches)
```

- How many words start and end with a vowel?

```python
matches = re.findall(r"^[AEIOUaeiou]([^AEIOUaeiou]*?[AEIOUaeiou])*?$", words, re.MULTILINE)
print(len(matches))
print(matches)
```

- How many words contain the same letter 3 times in a row?

```python
matches = re.findall(r"([A-Za-z])\1\1", words, re.MULTILINE)
print(len(matches))
print(matches)
```
